```python
top_supporters = support.groupby([
    "contributor_lastname",
    "contributor_firstname"
]).amount.sum().reset_index().sort_values(
    "amount", 
    ascending=False
).head(10)
```

```python
top_supporters.amount.plot.bar()
```

```python
chart = top_supporters.amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters["contributor_fullname"] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
top_supporters.to_csv("top_supporters.csv", index=False)
```

```python
top_opponents = oppose.groupby([
    "contributor_lastname",
    "contributor_firstname"
]).amount.sum().reset_index().sort_values(
    "amount", 
    ascending=False
).head(10)
```

```python
top_opponents.amount.plot.bar()
```

```python
top_opponents["contributor_fullname"] = top_opponents.contributor_firstname + " " + top_opponents.contributor_lastname
```

```python
top_opponents.amount.plot.barh()
```

```python
chart = top_opponents.amount.plot.barh()
chart.set_yticklabels(top_opponents.contributor_fullname)
```

```python
top_opponents.to_csv("top_opponents.csv", index=False)
```
